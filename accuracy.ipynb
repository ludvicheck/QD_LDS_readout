```python
import sys
#sys.path.append("\anaconda3\pkgs") 
from inputlds import*
from QD_LDS_functions import*
import numpy as np
from math import sqrt
import random
```

```python
num_trajectories = 50 #50
timesteps = 30

YC0 = generate_C0_data(num_trajectories, timesteps)
YC1 = generate_C1_data(num_trajectories, timesteps)
```

```python
systems = ["def_1_15", "def_1_30", "def_1_60", "def_1_200", "def_10_15", "def_10_30", "def_10_60", "def_10_200", "def_50_15", "def_50_30", "def_50_60", "def_50_200",
          "N4sid_1_15", "N4sid_1_30", "N4sid_1_60", "N4sid_1_200", "N4sid_10_15", "N4sid_10_30", "N4sid_10_60", "N4sid_10_200", "N4sid_50_15", "N4sid_50_30", "N4sid_50_60", "N4sid_50_200",
          "ssa_1_15", "ssa_1_30", "ssa_1_60", "ssa_1_200", "ssa_10_15", "ssa_10_30", "ssa_10_60", "ssa_10_200", "ssa_50_15", "ssa_50_30", "ssa_50_60", "ssa_50_200"]

decisionsC0 = []
for system in systems:
    system_decisionsC0 = []
    
    parts = system.split("_", maxsplit = 1)
    C0_string = "C0_".join(["_".join(parts[:-1]), parts[-1]])
    C1_string = "C1_".join(["_".join(parts[:-1]), parts[-1]])

    GC0 = pd.read_csv(f'data/{C0_string}_A.csv', sep=',', header=None)
    GC0 = np.array(GC0.values)
    FdashC0 = pd.read_csv(f'data/{C0_string}_C.csv', sep=',', header=None)
    FdashC0 = np.array(FdashC0.values)
    KC0 = pd.read_csv(f'data/{C0_string}_K.csv', sep=',', header=None)
    KC0 = np.array(KC0.values)
    ObsC0 = pd.read_csv(f'data/{C0_string}_noise.csv', sep=',', header=None)
    ObsC0 = np.array(ObsC0.values)
    ProcC0 = np.matmul(KC0, ObsC0)

    GC1 = pd.read_csv(f'data/{C1_string}_A.csv', sep=',', header=None)
    GC1 = np.array(GC1.values)
    FdashC1 = pd.read_csv(f'data/{C1_string}_C.csv', sep=',', header=None)
    FdashC1 = np.array(FdashC1.values)
    KC1 = pd.read_csv(f'data/{C1_string}_K.csv', sep=',', header=None)
    KC1 = np.array(KC1.values)
    ObsC1 = pd.read_csv(f'data/{C1_string}_noise.csv', sep=',', header=None)
    ObsC1 = np.array(ObsC1.values)
    ProcC1 = np.matmul(KC1, ObsC1)
    
    print(f'\n {system} on YC0 \n C0 matrices: {GC0}, {FdashC0}, {ObsC0}, {ProcC0} \n C1 matrices: {GC1}, {FdashC1}, {ObsC1}, {ProcC1}')
    
    for trajectory in YC0:

        preds0, errors0 = Kalman_filter(GC0, FdashC0.T, ProcC0, ObsC0, trajectory)
        preds1, errors1 = Kalman_filter(GC1, FdashC1.T, ProcC1, ObsC1, trajectory)

        decision = labelled_decission(errors0, errors1)

        system_decisionsC0.append(decision)
        
    print(f'system decisions on C0: {system_decisionsC0}')

    decisionsC0.append(system_decisionsC0)

#print(decisionsC0)

decisionsC1 = []

for system in systems:
    system_decisionsC1 = []
    
    parts = system.split("_", maxsplit = 1)
    C0_string = "C0_".join(["_".join(parts[:-1]), parts[-1]])
    C1_string = "C1_".join(["_".join(parts[:-1]), parts[-1]])

    GC0 = pd.read_csv(f'data/{C0_string}_A.csv', sep=',', header=None)
    GC0 = np.array(GC0.values)
    FdashC0 = pd.read_csv(f'data/{C0_string}_C.csv', sep=',', header=None)
    FdashC0 = np.array(FdashC0.values)
    KC0 = pd.read_csv(f'data/{C0_string}_K.csv', sep=',', header=None)
    KC0 = np.array(KC0.values)
    ObsC0 = pd.read_csv(f'data/{C0_string}_noise.csv', sep=',', header=None)
    ObsC0 = np.array(ObsC0.values)
    ProcC0 = np.matmul(KC0, ObsC0)

    GC1 = pd.read_csv(f'data/{C1_string}_A.csv', sep=',', header=None)
    GC1 = np.array(GC1.values)
    FdashC1 = pd.read_csv(f'data/{C1_string}_C.csv', sep=',', header=None)
    FdashC1 = np.array(FdashC1.values)
    KC1 = pd.read_csv(f'data/{C1_string}_K.csv', sep=',', header=None)
    KC1 = np.array(KC1.values)
    ObsC1 = pd.read_csv(f'data/{C1_string}_noise.csv', sep=',', header=None)
    ObsC1 = np.array(ObsC1.values)
    ProcC1 = np.matmul(KC1, ObsC1)
    
    print(f'\n {system} on YC1 \n C0 matrices: {GC0}, {FdashC0}, {ObsC0}, {ProcC0} \n C1 matrices: {GC1}, {FdashC1}, {ObsC1}, {ProcC1}')
    
    for trajectory in YC1:

        preds0, errors0 = Kalman_filter(GC0, FdashC0.T, ProcC0, ObsC0, trajectory)
        preds1, errors1 = Kalman_filter(GC1, FdashC1.T, ProcC1, ObsC1, trajectory)

        decision = labelled_decission(errors0, errors1)

        system_decisionsC1.append(decision)
        
    print(f'system decisions on C1: {system_decisionsC1}')

    decisionsC1.append(system_decisionsC1)

#print(decisionsC1)
    
```

```python
system_accuracies = []
system_precisions = []
system_recalls = []
system_F1scores = []

#for F1 score we consider identified state 1 as positive and state 0 as negative

for i in range(len(systems)):
    system_accuracy = ((len(decisionsC0[i]) - sum(decisionsC0[i]) + sum(decisionsC1[i])) / (len(decisionsC0[i]) + len(decisionsC1[i]))) * 100
    system_accuracies.append(system_accuracy)
    
    system_precision = sum(decisionsC1[i]) / ((len(decisionsC0[i]) - sum(decisionsC0[i])) + sum(decisionsC1[i]))
    system_precisions.append(system_precision)
    
    if (sum(decisionsC0[i]) + sum(decisionsC1[i])) == 0:
        system_recall = 0
    else:
        system_recall = sum(decisionsC1[i]) / (sum(decisionsC0[i]) + sum(decisionsC1[i]))
    system_recalls.append(system_recall)
    
    if (system_precision + system_recall) == 0:
        system_F1score = 0
    else:   
        system_F1score = 2 * ((system_precision * system_recall) / (system_precision + system_recall))                    
    system_F1scores.append(system_F1score)
```

```python
for i in range(len(systems)):
    print(f'{systems[i]} accuracy: {system_accuracies[i]}%, precision: {system_precisions[i]}, recall: {system_recalls[i]}, F1 score: {system_F1scores[i]}')
```
